```python
from biomart import BiomartServer
import pandas as pd
```

```python
def build_entrez_ensembl_translation_dataframe(gene_list, path):
    
    server = BiomartServer( "http://uswest.ensembl.org/biomart" )
    ds = server.datasets['hsapiens_gene_ensembl']
    
    df_rows = []
    response = ds.search({
        'filters': {
            'entrezgene_id': gene_list[0:10000]
        },
    'attributes': ['entrezgene_id', 'ensembl_peptide_id', 'ensembl_gene_id'] 
        }, header=1)
    
    
    for line in response.iter_lines():
        df_rows.append(line.decode('utf-8').split('\t'))
    
    
    df = pd.DataFrame(df_rows)
    df = df.rename(columns=df.iloc[0]).drop(df.index[0])
    
    
    df.to_csv(path+"../translation_df.csv")
    return df
```

```python
path="/home/colombelli/Documents/datasets/graph-omics/KIRC/"
gene_cols = pd.read_csv(path+'gene_proc.csv', index_col=0, nrows=0).columns.tolist()
gene_list = [i.split("|")[1] for i in gene_cols[:-1]]
translation_df = entrez_to_ensembl_dict = build_entrez_ensembl_translation_dataframe(gene_list, path)
```
